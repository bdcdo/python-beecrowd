# A função `print()`

O `print()` exibe informações na tela. É o comando mais básico do Python.

```python
print("Hello World!")
print(42)
print("O resultado é", 3 + 5)
```

## Variáveis

Variáveis guardam valores para usar depois. Em Python, basta escolher um nome e atribuir com `=`.

Existem 3 tipos que vamos usar:
- **int** — número inteiro: `10`, `-3`
- **float** — número decimal: `3.14`, `-0.5`
- **str** — texto: `"Olá"`, `"Python"`

```python
nome = "Maria"
idade = 25
nota = 8.5

print(nome)
print(idade)
print(nota)
```
