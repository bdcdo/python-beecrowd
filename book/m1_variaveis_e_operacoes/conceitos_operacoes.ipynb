# Operadores aritméticos

Python suporta os seguintes operadores:

| Operador | Operação | Exemplo | Resultado |
|---|---|---|---|
| `+` | Adição | `5 + 3` | `8` |
| `-` | Subtração | `10 - 4` | `6` |
| `*` | Multiplicação | `7 * 3` | `21` |
| `/` | Divisão | `7 / 2` | `3.5` |
| `//` | Divisão inteira | `7 // 2` | `3` |
| `%` | Resto | `7 % 2` | `1` |
| `**` | Potência | `2 ** 3` | `8` |

Na dúvida sobre a ordem das operações, use **parênteses**.

```python
a = 15
b = 4

print(a + b)    # 19
print(a - b)    # 11
print(a * b)    # 60
print(a / b)    # 3.75
print(a // b)   # 3  (parte inteira)
print(a % b)    # 3  (resto)
print(a ** b)   # 50625  (15 elevado a 4)
```

## Divisão inteira e resto — padrão para cédulas e tempo

Os operadores `//` (divisão inteira) e `%` (resto) formam uma dupla muito útil:
- `//` responde: **quantas vezes cabe?**
- `%` responde: **quanto sobra?**

Esse padrão aparece em problemas de cédulas (1018) e conversão de tempo (1019).

```python
# Simulando entrada: valor = 576
valor = 576

# Quantas notas de 100?
n100 = valor // 100      # 5
resto = valor % 100       # 76

# Quantas notas de 50?
n50 = resto // 50         # 1
resto = resto % 50        # 26

print(f"100: {n100}")
print(f"50: {n50}")
print(f"Resto: {resto}")
```

## Módulo `math`

O módulo `math` oferece funções matemáticas prontas.
Para usá-lo, escreva `import math` no início do código.

| Função | Descrição | Exemplo |
|---|---|---|
| `math.sqrt(x)` | Raiz quadrada | `math.sqrt(25)` → `5.0` |
| `math.pi` | Constante π | `3.141592…` |

```python
import math

# Raiz quadrada
print(math.sqrt(25))   # 5.0

# Área do círculo (problema 1002)
R = 2.0
area = math.pi * R ** 2
print(f"A={area:.4f}")  # A=12.5664
```

## Formatando a saída com f-string

Uma **f-string** permite embutir variáveis diretamente no texto.
Basta colocar `f` antes das aspas e usar `{variável}` dentro delas.

Para controlar as casas decimais, use `{variável:.Nf}` onde `N` é o número de casas.

```python
nome = "João"
X = 10
Y = 15
soma = X + Y

# Embutir variáveis no texto
print(f"SOMA = {soma}")
print(f"{nome} tem {X} anos")

# Casas decimais
area = 3.141592 * 5.0 ** 2
print(f"A={area:.4f}")
```

```python
# Exemplo prático — Beecrowd 1001 (Soma Simples)
A = 30
B = 10
SOMA = A + B
print(f"SOMA = {SOMA}")
```

## Fatiamento de strings

Podemos extrair pedaços de um texto usando a sintaxe `texto[início:fim]`.
O caractere na posição `início` é incluído; o da posição `fim` **não** é incluído.

Esse padrão é útil para formatar CPF (problema 2763), datas, etc.

```python
cpf = "12345678900"

# Fatiar: texto[início:fim]
parte1 = cpf[0:3]    # "123"
parte2 = cpf[3:6]    # "456"
parte3 = cpf[6:9]    # "789"
parte4 = cpf[9:11]   # "00"

print(f"{parte1}.{parte2}.{parte3}-{parte4}")
```

## Resumo

| Conceito | Sintaxe | Exemplo |

| Raiz quadrada | `math.sqrt(x)` | `math.sqrt(25)` → `5.0` |
| Constante π | `math.pi` | `3.141592…` |
| f-string | `f"texto {var}"` | `f"SOMA = {s}"` |
| Casas decimais | `f"{var:.Nf}"` | `f"{area:.4f}"` |
| Fatiamento | `texto[a:b]` | `cpf[0:3]` → `"123"` |
