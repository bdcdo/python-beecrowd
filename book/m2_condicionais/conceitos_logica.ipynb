# Operadores lógicos: and, or, not

Podemos combinar várias condições em um mesmo `if` usando operadores lógicos:

- **`and`** — o resultado é `True` somente se **ambas** as condições forem verdadeiras.
- **`or`** — o resultado é `True` se **pelo menos uma** das condições for verdadeira.
- **`not`** — **inverte** o valor: `True` vira `False` e vice-versa.

```python
idade = 20
renda = 3000

# and — as duas condições precisam ser verdadeiras
if idade >= 18 and renda >= 2000:
    print("Aprovado")

# or — pelo menos uma precisa ser verdadeira
nota1 = 4
nota2 = 8
if nota1 >= 7 or nota2 >= 7:
    print("Passou em pelo menos uma prova")
```

## Paridade com `%`

O operador `%` (módulo) retorna o **resto da divisão inteira**. Ele é muito útil para verificar paridade e múltiplos:

- `x % 2 == 0` — `x` é **par**
- `x % 2 != 0` — `x` é **ímpar**
- `x % n == 0` — `x` é **múltiplo de n**

```python
n = 15
if n % 2 == 0:
    print("Par")
else:
    print("Ímpar")

if n % 3 == 0:
    print(f"{n} é múltiplo de 3")
```

## Condicionais aninhados

Podemos colocar um `if` **dentro** de outro `if`, criando uma árvore de decisão. A cada nível, a indentação aumenta. Esse padrão é útil quando a segunda decisão depende do resultado da primeira.

```python
# Beecrowd 1049 — Animal
tipo = "vertebrado"
classe = "ave"

if tipo == "vertebrado":
    if classe == "ave":
        print("aguia")
    elif classe == "mamifero":
        print("vaca")
    elif classe == "reptil":
        print("cobra")
else:
    if classe == "inseto":
        print("pulga")
    elif classe == "anelideo":
        print("minhoca")
```

## Aritmética modular — Tempo de Jogo

O operador `%` também resolve cálculos que envolvem **ciclos**, como horas (ciclo de 24 h). O truque é usar `(fim - inicio) % 24`. Se o resultado for `0`, significa que início e fim são iguais, então a duração é **24 horas** (o jogo não pode durar 0 h).

```python
# Beecrowd 1046 — Tempo de Jogo
inicio = 16
fim = 2

duracao = (fim - inicio) % 24
if duracao == 0:
    duracao = 24

print(f"O JOGO DUROU {duracao} HORA(S)")
```

## Cálculo por faixas — Imposto de Renda

No imposto progressivo, cada **faixa** do salário é taxada com uma alíquota diferente. Não se aplica a alíquota mais alta sobre o salário inteiro — apenas sobre a parcela que cai naquela faixa.

```python
# Beecrowd 1051 — Imposto de Renda (simplificado)
salario = 4500.00
imposto = 0.0

if salario > 4000.00:
    imposto += (salario - 4000.00) * 0.08
    salario_calc = 4000.00
else:
    salario_calc = salario

if salario_calc > 3000.00:
    imposto += (salario_calc - 3000.00) * 0.08
    salario_calc = 3000.00

if salario_calc > 2000.00:
    imposto += (salario_calc - 2000.00) * 0.08

if imposto > 0:
    print(f"R$ {imposto:.2f}")
else:
    print("Isento")
```

## `math.sqrt` em condicionais — Bhaskara

A função `math.sqrt()` calcula a raiz quadrada, mas ela **não aceita números negativos**. Por isso, antes de usá-la, é obrigatório verificar se o delta é maior ou igual a zero. Também é preciso garantir que `a` não seja zero, para evitar divisão por zero.

```python
import math

# Beecrowd 1036 — Bhaskara
a, b, c = 2.0, -7.0, 3.0

delta = b**2 - 4*a*c

if a == 0:
    print("Impossivel calcular")
elif delta < 0:
    print("Impossivel calcular")
else:
    x1 = (-b + math.sqrt(delta)) / (2*a)
    x2 = (-b - math.sqrt(delta)) / (2*a)
    print(f"R1 = {x1:.5f}")
    print(f"R2 = {x2:.5f}")
```
