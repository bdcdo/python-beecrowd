## if simples

O comando `if` testa uma condição. Se ela for **verdadeira** (`True`), o bloco de código indentado abaixo é executado. Se for **falsa** (`False`), o bloco é simplesmente ignorado.

```python
idade = 18
if idade >= 18:
    print("Maior de idade")
```

## if/else

Quando queremos executar um bloco caso a condição seja verdadeira **e outro bloco** caso seja falsa, usamos `else`. O `else` funciona como um "senão": se a condição do `if` não for satisfeita, o código dentro do `else` é executado.

```python
numero = 7
if numero % 2 == 0:
    print("Par")
else:
    print("Ímpar")
```

## if/elif/else

Quando existem **mais de duas opções**, usamos `elif` (abreviação de "else if"). As condições são testadas de cima para baixo: assim que uma for verdadeira, seu bloco é executado e **todos os demais são ignorados**. O `else` final é opcional e funciona como um caso padrão.

```python
# Beecrowd 1038 — Lanche
codigo = 3

if codigo == 1:
    print(f"R$ {4.00:.2f}")
elif codigo == 2:
    print(f"R$ {4.50:.2f}")
elif codigo == 3:
    print(f"R$ {5.00:.2f}")
elif codigo == 4:
    print(f"R$ {2.00:.2f}")
elif codigo == 5:
    print(f"R$ {1.50:.2f}")
```

```python
# Beecrowd 2344 — Notas da Prova
nota = 7

if nota >= 9:
    print("A")
elif nota >= 7:
    print("B")
elif nota >= 5:
    print("C")
elif nota >= 3:
    print("D")
else:
    print("E")
```

## Operadores de comparação

Os operadores de comparação comparam dois valores e retornam `True` ou `False`.

| Operador | Significado       | Exemplo  | Resultado |
|----------|-------------------|----------|-----------|
| `==`     | Igual a           | `5 == 5` | `True`    |
| `!=`     | Diferente de      | `5 != 3` | `True`    |
| `<`      | Menor que         | `3 < 5`  | `True`    |
| `>`      | Maior que         | `5 > 3`  | `True`    |
| `<=`     | Menor ou igual a  | `5 <= 5` | `True`    |
| `>=`     | Maior ou igual a  | `5 >= 6` | `False`   |

```python
a = 10
b = 20
print(a == b)    # False
print(a != b)    # True
print(a < b)     # True
print(a >= 10)   # True
```

## Erro comum: `=` vs `==`

Um dos erros mais frequentes é confundir **atribuição** com **comparação**:

- `=` (um sinal) **atribui** um valor a uma variável: `x = 5`
- `==` (dois sinais) **compara** dois valores: `x == 5`
