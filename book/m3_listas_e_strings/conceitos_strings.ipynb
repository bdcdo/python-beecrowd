# Strings em Python

## 1. Strings — indexação e `len()`

Strings são **sequências de caracteres**. Cada caractere tem uma posição (índice), começando em `0`.  
Usamos `len()` para saber o **tamanho** da string.

```python
texto = "Python"
print(texto[0])      # P
print(texto[5])      # n
print(len(texto))    # 6
```

## 2. Fatiamento (slicing)

Podemos extrair **partes** de uma string com a sintaxe `[início:fim]`.  
O caractere na posição `fim` **não é incluído**. Se omitirmos o início, começa do `0`.

```python
# Beecrowd 2760 — primeiros 10 caracteres
nome = "Julieta Capuleto"
print(nome[:10])     # "Julieta Ca"

# Beecrowd 1864 — extrair partes de data DD/MM/AAAA
data = "23/02/2026"
dia = data[0:2]      # "23"
mes = data[3:5]      # "02"
ano = data[6:10]     # "2026"
print(f"{dia}/{mes}/{ano}")
```

## 3. `len()` para validação

Usamos `len()` para verificar se um texto **cabe dentro de um limite** de caracteres.

```python
# Beecrowd 2165 — Tuitando (limite 140 caracteres)
tweet = "Estudando Python para resolver exercícios!"

if len(tweet) <= 140:
    print("TWEET")
else:
    print("MUTE")
```

## 4. `chr()` e `ord()` — tabela ASCII

Cada caractere possui um **código numérico** na tabela ASCII.  
- `chr(n)` converte um número no caractere correspondente  
- `ord(c)` converte um caractere no seu número

```python
# Tabela ASCII: A=65, B=66, ... Z=90
print(chr(65))       # A
print(ord("A"))      # 65

# Beecrowd 2727 — Código Secreto
# Somar 3 ao código de cada letra
letra = "A"
codigo = ord(letra) + 3
print(chr(codigo))   # D
```

## 5. Concatenação — juntar strings com `+`

Usamos o operador `+` para **juntar** (concatenar) duas ou mais strings em uma só.

```python
primeiro = "Maria"
segundo = "Silva"
nome_completo = primeiro + " " + segundo
print(nome_completo)   # Maria Silva

# Construindo saída caractere a caractere
resultado = ""
for c in "ABC":
    resultado = resultado + c + "."
print(resultado)       # A.B.C.
```
