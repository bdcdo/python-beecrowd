# Introdução a Python com Beecrowd

Bem-vindo ao **Introdução a Python com Beecrowd** — um material de aula interativo que combina explicações de conceitos Python com exercícios práticos da plataforma [Beecrowd](https://judge.beecrowd.com/).

## O que você vai aprender

Este guia cobre os 4 primeiros módulos de programação em Python:

| Módulo | Tema | Exercícios |
|--------|------|:----------:|
| **M1** | Variáveis e Operações | 40 |
| **M2** | Condicionais | 25 |
| **M3** | Listas e Strings | 16 |
| **M4** | Loops | 35 |

**Total: 116 exercícios** organizados por nível de dificuldade.

## Estrutura de cada módulo

Cada módulo contém:

1. **Notebooks de conceitos** — explicações com exemplos executáveis
2. **Listas de exercícios** — problemas Beecrowd classificados em 3 níveis:
   - **Básico**: conceitos fundamentais, problemas diretos
   - **Fixação**: prática adicional, pequenas variações
   - **Aprofundamento**: combinação de conceitos, problemas mais elaborados

## Pré-requisitos

- Python 3.8 ou superior instalado
- Conta na plataforma [Beecrowd](https://judge.beecrowd.com/)
- Vontade de praticar!

## Como começar

Navegue pelo menu lateral e comece pelo **Módulo 1**. Recomendamos seguir a ordem: primeiro leia os notebooks de conceitos, depois resolva os exercícios do nível básico antes de avançar para os próximos níveis.
