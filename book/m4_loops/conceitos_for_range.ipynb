# Loops com `for` e `range()`

O comando `for` repete um bloco de código um número determinado de vezes.
Combinado com `range()`, controlamos exatamente quantas repetições acontecem e quais valores são gerados.

## 1. `for` com `range()`

A função `range(start, stop, step)` gera uma sequência de números:
- **start**: onde começa (incluso, padrão 0)
- **stop**: onde termina (excluso)
- **step**: o incremento entre valores (padrão 1)

Exemplo: imprimir os números pares de 2 a 10.

```python
# Beecrowd 1059 — Números Pares (simplificado)
for numero in range(2, 11, 2):
    print(numero)
```

## 2. Acumulador — somando valores

Um **acumulador** é uma variável que começa em `0` e vai recebendo valores com `+=` a cada repetição.
É o padrão mais comum em loops: somar uma sequência de números.

```python
# Beecrowd 1071 — Soma de Ímpares Consecutivos (simplificado)
# Somar os ímpares entre 1 e 10
soma = 0

for numero in range(1, 11):
    if numero % 2 != 0:
        soma = soma + numero

print(f"Soma dos ímpares: {soma}")
```

## 3. Acumulador de produto — fatorial

Quando queremos multiplicar valores em sequência, o acumulador começa em `1` (elemento neutro da multiplicação) e usamos `*=`.
O exemplo clássico é o **fatorial**: `5! = 1 × 2 × 3 × 4 × 5 = 120`.

```python
# Beecrowd 1153 — Fatorial Simples
n = 5

fatorial = 1
for i in range(1, n + 1):
    fatorial = fatorial * i

print(f"{n}! = {fatorial}")
```

## 4. Contador

Um **contador** é uma variável que começa em `0` e aumenta em `1` a cada vez que uma condição é satisfeita.
A diferença para o acumulador: o acumulador soma *valores*, o contador soma *ocorrências*.

```python
# Beecrowd 1060 — Números Positivos (simplificado)
valores = [8, -3, 7, -1, 5, -2]

contador = 0

for i in range(6):
    if valores[i] > 0:
        contador = contador + 1

print(f"{contador} valores positivos")
```

## 5. Acumulador + Contador = Média

Combinando os dois padrões: o **acumulador** guarda a soma e o **contador** guarda quantos valores entraram na soma. No final, dividimos um pelo outro para obter a média.

```python
# Beecrowd 1064 — Positivos e Média (simplificado)
valores = [8.0, -3.0, 7.0, -1.0, 5.0, -2.0]

contador = 0
soma = 0.0

for i in range(6):
    if valores[i] > 0:
        contador = contador + 1
        soma = soma + valores[i]

print(f"{contador} valores positivos")
print(f"Média: {soma / contador:.1f}")
```

## 6. Rastreando o maior/menor

Para encontrar o **maior valor** de uma sequência sem usar lista, guardamos o primeiro valor como candidato e comparamos com cada novo valor.
Se o novo valor for maior, atualizamos o candidato e registramos a posição.

```python
# Beecrowd 1080 — Maior e Posição (simplificado)
valores = [15, 42, 8, 73, 31, 56, 19, 64, 27, 88]

maior = valores[0]
posicao = 1

for i in range(1, len(valores)):
    if valores[i] > maior:
        maior = valores[i]
        posicao = i + 1  # posição começa em 1

print(f"Maior: {maior}")
print(f"Posição: {posicao}")
```

## 7. Resumo

| Padrão | Inicialização | Atualização | Uso típico |
|--------|---------------|-------------|------------|
| **Acumulador (soma)** | `soma = 0` | `soma += valor` | Soma de sequências |
| **Acumulador (produto)** | `resultado = 1` | `resultado *= valor` | Fatorial, potências |
| **Contador** | `contador = 0` | `contador += 1` (dentro de `if`) | Contar ocorrências |
| **Acumulador + Contador** | Ambos acima | Ambos acima | Calcular média |
| **Rastreamento** | `melhor = primeiro_valor` | `if novo > melhor: melhor = novo` | Maior, menor, posição |
