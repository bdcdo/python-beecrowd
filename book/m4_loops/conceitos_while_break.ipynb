# Loops com `while` e `break`

O `while` repete um bloco **enquanto** uma condição for verdadeira.
Diferente do `for`, não sabemos de antemão quantas vezes o loop vai executar — ele para quando a condição muda.

## 1. `while True` + `break`

O padrão `while True` cria um loop infinito que só para com `break`.
Útil quando queremos **repetir até** que algo aconteça — por exemplo, até o usuário acertar uma senha.

```python
# Beecrowd 1114 — Senha Fixa (simplificado)
tentativas = [1234, 5678, 2002]  # simulando input

for tentativa in tentativas:
    senha = tentativa
    if senha == 2002:
        print("Acesso Permitido")
        break
    else:
        print("Senha Invalida")
```

## 2. `while` com condição direta

Quando a condição de parada é clara desde o início, escrevemos direto no `while`.
Exemplo: uma cidade A cresce até ultrapassar a população da cidade B.

```python
# Beecrowd 1160 — Crescimento Populacional (simplificado)
pa = 100    # população A
pb = 200    # população B
g1 = 10     # crescimento A (%)
g2 = 2      # crescimento B (%)

anos = 0
while pa <= pb:
    pa = pa + int(pa * g1 / 100)
    pb = pb + int(pb * g2 / 100)
    anos = anos + 1

print(f"{anos} anos para A ultrapassar B")
```

## 3. Validação de entrada

Um uso muito comum de `while` é **validar dados**: repetir a leitura até que o valor esteja dentro do intervalo esperado.
Exemplo: pedir uma nota entre 0 e 10, rejeitando valores fora da faixa.

```python
# Beecrowd 1117 — Validação de Nota (simplificado)
entradas = [15.0, -3.0, 7.5]  # simulando input

i = 0
while True:
    nota = entradas[i]
    i = i + 1
    if nota < 0 or nota > 10:
        print("nota invalida")
    else:
        break

print(f"Nota aceita: {nota:.2f}")
```

## 4. Loops aninhados

Um loop dentro de outro. Para cada repetição do loop **externo**, o loop **interno** executa completamente.
Útil para gerar combinações ou preencher tabelas.

```python
# Beecrowd 1095 — Sequência IJ (simplificado)
# Para cada I de 1 a 3, J vai de 7 a 9
for i in range(1, 4):
    for j in range(7, 10):
        print(f"I={i} J={j}")
```

## 5. Combinando `while` e `for`

É comum ter um `while True` no nível externo (para ler vários casos de teste) e um `for` interno (para processar cada caso).
O `while` para com `break` quando uma condição de fim é encontrada.

```python
# Beecrowd 1101 — Sequência de Números e Soma (simplificado)
casos = [(2, 5), (3, 7), (0, 0)]  # simulando input

for m, n in casos:
    if m <= 0 or n <= 0:
        break

    menor = min(m, n)
    maior = max(m, n)

    soma = 0
    for i in range(menor, maior + 1):
        soma = soma + i

    print(f"De {menor} a {maior}: soma = {soma}")
```

## 6. Resumo

| Estrutura | Quando usar | Exemplo |
|-----------|-------------|----------|
| `for i in range(n)` | Repetir **n** vezes, sabendo de antemão | Tabuada, sequências fixas |
| `while True` + `break` | Repetir **até** algo acontecer | Senha, leitura até sentinela |
| `while condição` | Repetir **enquanto** condição for verdadeira | Crescimento, convergência |
| Loops aninhados | Gerar **combinações** ou processar tabelas | Sequências IJ, matrizes |
