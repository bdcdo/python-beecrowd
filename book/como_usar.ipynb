## Sobre o `input()` nos notebooks

```{admonition} Importante
:class: warning

Nos notebooks deste guia, **não usamos `input()` diretamente** porque ele bloqueia a execução automática das células. Em vez disso, simulamos a entrada com **atribuição direta**.
```

**No Beecrowd (código real):**

## Receitas de input() para o Beecrowd

Copie e cole o padrão que se encaixa no seu problema:

**Ler 1 número inteiro:**
```python
N = int(input())
```

**Ler 1 número decimal:**
```python
X = float(input())
```

**Ler 1 texto:**
```python
S = input()
```

**Ler 2 inteiros na mesma linha:**
```python
A, B = map(int, input().split())
```

**Ler 3 decimais na mesma linha:**
```python
A, B, C = map(float, input().split())
```

**Ler 3 inteiros na mesma linha:**
```python
A, B, C = map(int, input().split())
```

## Entendendo `split()` e `map()`

As receitas acima usam `split()` e `map()`. Vamos entender o que cada um faz.

### `split()` — separar texto em partes

O método `split()` pega uma string e **separa em uma lista**, usando espaços como divisor:

````python
"10 20 30".split()
````

O resultado é uma lista de strings: `['10', '20', '30']`. Note que os números ainda são **texto**, não números de verdade.

### `map()` — aplicar uma função a cada elemento

O `map()` recebe uma função e uma lista, e **aplica a função a cada elemento**. Usamos `map(int, ...)` para converter cada string em número inteiro:

````python
list(map(int, ['10', '20', '30']))
````

Agora temos números de verdade: `[10, 20, 30]`.

### O padrão completo

Quando o Beecrowd diz "leia dois inteiros na mesma linha", a entrada vem assim: `10 20`. O padrão completo combina tudo:

```python
A, B = map(int, input().split())
```

Passo a passo:
1. `input()` lê a linha inteira como texto: `"10 20"`
2. `.split()` separa por espaços: `['10', '20']`
3. `map(int, ...)` converte cada um para inteiro: `10, 20`
4. `A, B = ...` guarda cada valor em uma variável
